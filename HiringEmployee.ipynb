Importing all the Dependendencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from word2number import w2n
from sklearn.metrics import accuracy_score
```

EMPLOYEE SALARY PREDICTION

```python
main_data = pd.read_csv('hiring.csv')
main_data
```

Data Preprocessing

```python
main_data.experience = main_data.experience.fillna('zero')
median_test_score = math.floor(main_data['test_score(out of 10)'].median())
print("Median is : ",median_test_score)
main_data['test_score(out of 10)'] = main_data['test_score(out of 10)'].fillna(median_test_score)
main_data
```

Labelling Data with its Features and Target

```python
X = main_data[['experience','test_score(out of 10)','interview_score(out of 10)']]
y = main_data['salary($)']
```

```python
X.experience = X.experience.apply(w2n.word_to_num) # Words to Number Conversion
print(y)
X.head()
```

Splitting Labeled Dataset into Training and Testing

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=3,train_size=0.8)
```

```python
X_test = X_test.astype('int')
X_train = X_train.astype('int')
y_test = y_test.astype('int')
y_train = y_train.astype('int')
X_test,y_test
```

Model Selection and Training the Data to the models

```python
model_linear = linear_model.LinearRegression()
model_logistic = linear_model.LogisticRegression()
```

```python
model_linear.fit(X_train,y_train)
```

```python
model_logistic.fit(X_train,y_train)
```

```python
prediction = model_logistic.predict(X_test)
print(prediction)
accuracy = accuracy_score(y_test, prediction)
print(accuracy)
```
