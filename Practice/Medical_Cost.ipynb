Importing all Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, accuracy_score
```

Loading DataSet

```python
Cost_Data = pd.read_csv('Medical_Cost.csv')
Cost_Data.head()
```

```python
# Checking if there are any Null Values
Cost_Data.isnull().sum()
```

Data Preprocessing...

```python
Cost_Data.replace({'sex' : {'male':1,'female':0},
                   'smoker': {'yes':1,'no':0},
                   'region': {'southwest':0,'southeast':1,'northwest':2,'northeast':3}
                   }, inplace=True)
Cost_Data
```

```python
X = Cost_Data.drop(columns='medical charges',axis=1)
y = Cost_Data['medical charges']
```

Splitt into TRAIN and TEST

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=7)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predictions = model.predict(X_test)
accuracy = r2_score(y_test, y_predictions)
print(accuracy)
```
